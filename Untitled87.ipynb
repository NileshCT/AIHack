```python
from bs4 import BeautifulSoup
import requests
```

```python
def fetch_website_content(url):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, "html.parser")
            # Extract meaningful text (modify selectors as needed)
            content = soup.get_text(separator="\n", strip=True)
            return content
        else:
            print(f"Failed to fetch URL: {response.status_code}")
            return None
    except Exception as e:
        print(f"Error fetching URL: {e}")
        return None

url = "https://github.com/aws-samples/aws-healthimaging-samples"
website_content = fetch_website_content(url)
if website_content:
    print("Website Content Extracted Successfully")
```

```python
def preprocess_content(content):
    # Remove excessive whitespace
    cleaned_content = " ".join(content.split())
    # Truncate or summarize if too long
    if len(cleaned_content) > 2000:  # Limit length for Hugging Face API
        cleaned_content = cleaned_content[:2000] + "..."
    return cleaned_content

processed_content = preprocess_content(website_content)
```

```python
prompt = f"""
Generate a KYC document based on the following website content:

Website Content:
{processed_content}

The content should be precious and it should be perfect format.
"""
```

```python
import requests

# Hugging Face API details
API_KEY = "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"
MODEL_URL = "https://api-inference.huggingface.co/models/gpt2"  # Using a known working model

prompt = f"""
Generate a KYC document based on the following website content:

Website Content:
{processed_content}

The content should be precious and it should be perfect format.
"""

# API headers with the `Accept` header to request plain text
headers = {
    "Authorization": f"Bearer {API_KEY}",
    "Accept": "text/plain"  # Request response in plain text format
}

# Payload for the request
payload = {
    "inputs": prompt,
    "parameters": {
        "max_length": 500,
        "temperature": 0.7,
        "top_p": 0.9,
    }
}

# Sending the POST request
response = requests.post(MODEL_URL, headers=headers, json=payload)

# Handling the response as raw text
if response.status_code == 200:
    # Directly extract the text from the response
    kyc_document = response.text  # Raw text response
    print("Generated KYC Document:")
    print(kyc_document)
else:
    print(f"Error: {response.status_code} - {response.text}")
```

```python
with open("Generated_KYC_Document.txt", "w") as file:
    file.write(kyc_document)
print("KYC Document saved successfully.")
```
